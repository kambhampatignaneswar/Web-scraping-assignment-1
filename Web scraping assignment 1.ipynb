```python
import pandas as pd
import requests
import csv
from bs4 import BeautifulSoup
```

```python
url = "https://www.flipkart.com/search?q=iphone"
```

```python
req = requests.get(url)
```

```python
req
```

```python
soup = BeautifulSoup(req.content,'html.parser')
```

```python
soup
```

```python
titles = soup.find_all('div',class_='_4rR01T')
print(titles)
```

```python
ratings = soup.find_all('div',class_='_3LWZlK')
print(ratings)
```

```python
price = soup.find_all('div',class_='_30jeq3 _1_WHN1')
print(price)
```

```python
mt=[]
mr=[]
mp=[]
```

```python
for titles,ratings,price in zip(titles,ratings,price):
    mt.append(titles.text)
    mr.append(ratings.text)
    mp.append(price.text)
```

```python
print(mp)
```

```python
print(mt)
```

```python
print(mr)
```

```python
d={'mt':mt,'mr':mr,}
```

```python
model=pd.DataFrame(data=d)
```

```python
model.to_csv("iphonedata.csv")
```
